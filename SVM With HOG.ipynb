```python
import os
import cv2
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from skimage.feature import hog

# Set the path to your dataset directory
dataset_dir = "H:/15 classes 6k Images"

# Set the HOG parameters
orientations = 9
pixels_per_cell = (8, 8)
cells_per_block = (2, 2)

# List to store HOG features and labels
hog_features = []
labels = []
class_names = []

# Loop through each class folder
for class_name in os.listdir(dataset_dir):
    class_dir = os.path.join(dataset_dir, class_name)
    class_index = len(class_names)  # Assign a unique index to each class
    
    # Add the class name to the list of class names
    class_names.append(class_name)
    
    # Loop through each image in the class folder
    for image_name in os.listdir(class_dir):
        image_path = os.path.join(class_dir, image_name)
        try:
            # Read the image
            image = cv2.imread(image_path)
            
            # Convert the image to grayscale
            gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
            
            # Extract HOG features
            features = hog(gray, orientations=orientations, pixels_per_cell=pixels_per_cell, cells_per_block=cells_per_block)
            
            # Append the features and label to the lists
            hog_features.append(features)
            labels.append(class_index)
        except Exception as e:
            print(f"Error processing image: {image_path}")
            print(e)
```

```python
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
import numpy as np

# Convert the HOG features and labels to numpy arrays
hog_features = np.array(hog_features)
labels = np.array(labels)

# Split the data into training, testing, and validation sets
X_train, X_test, y_train, y_test = train_test_split(hog_features, labels, test_size=0.3, random_state=42)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size=0.33, random_state=42)

# Apply PCA for dimensionality reduction
n_components_list = [10, 50, 100, 200]  # Example list of n_components values
for n in n_components_list:
    pca = PCA(n_components=n)
    X_train_pca = pca.fit_transform(X_train)
pca = PCA(n_components=100)  # Adjust the number of components as needed
X_train_pca = pca.fit_transform(X_train)
X_val_pca = pca.transform(X_val)
X_test_pca = pca.transform(X_test)

# Create an SVM classifier
svm = SVC()

# Train the SVM classifier
svm.fit(X_train_pca, y_train)

# Make predictions on the validation set
y_pred = svm.predict(X_val_pca)

# Calculate the accuracy of the model on the validation set
accuracy = accuracy_score(y_val, y_pred)

# Print the accuracy
print("Validation Accuracy:", accuracy)

# Make predictions on the test set
y_pred_test = svm.predict(X_test_pca)

# Calculate the accuracy of the model on the test set
accuracy_test = accuracy_score(y_test, y_pred_test)

# Print the test accuracy
print("Test Accuracy:", accuracy_test)
```
